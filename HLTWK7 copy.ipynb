```python
import altair as alt
```

```python
z1['type'].replace(1, 'Mammals', inplace = True)
z1['type'].replace(2, 'Birds', inplace = True)
z1['type'].replace(3, 'Reptile', inplace = True)
z1['type'].replace(4, 'Fish', inplace = True)
z1['type'].replace(5, 'Amphibians', inplace = True)
z1['type'].replace(6, 'Insects', inplace = True)
z1['type'].replace(7, 'Others', inplace = True)
```

```python
z1.sample()
```

```python
alt.Chart(z1).mark_bar().encode(
    x = alt.X('legs', bin = True),
    y ='count()'
)
```

```python
pred = alt.Chart(z1).mark_point().encode(
    x = alt.X('type'),
    y ='type',
    color = 'predator'
)
```

```python
legs = alt.Chart(z1).mark_bar().encode(
    x = alt.X('legs'),
    y ='count()',
    color = 'type'
)
```

```python
legs | pred
```

```python
z1.loc[z1.legs == 5]
```

I decided to choose this design choice to see how many legs each type of animal had. I was then curious to find out which animal had 5 legs so I included the above code using the loc function.
I then wanted to know the propensity of predators amongst the animals so I created a scatter plot chart to show this.
